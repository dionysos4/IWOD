### This script splits the Kitti validation set into validation and test
it seperates the data, that validation and test data come from different sequences

```python
import numpy as np
import os
```

```python
MAPPINGPATH = "<yout path to ../mapping>/mapping"
TRAINRANDPATH = os.path.join(MAPPINGPATH, "train_rand.txt")
TRAINMAPPINGPATH = os.path.join(MAPPINGPATH, "train_mapping.txt")
CONFIGPATH = "<your_path_to_config_dir>"
VALPATH = os.path.join(CONFIGPATH, "val.txt")
```

```python
# read train_rand.txt
with open(TRAINRANDPATH, 'r') as f:
    train_rand = f.readlines()
train_rand = train_rand[0].split(",")
train_rand = np.array(train_rand, dtype=np.int32)

# read train_mapping.txt
with open(TRAINMAPPINGPATH, 'r') as f:
    train_mapping = f.readlines()

# read val file
with open(VALPATH, 'r') as f:
    val_idx = f.readlines()
val_idx = [int(i.strip()) for i in val_idx]
```

```python
sequence_list = []
for idx in val_idx:
    line_num = train_rand[idx]-1
    sequence_list.append(train_mapping[line_num].strip())

# use only the first two strings from sequence list which are seperated by a space
sequence_list_split = [i.split(" ")[0]+" "+i.split(" ")[1]  for i in sequence_list]

# create dictionary with sequence_list as keys and the image number as values
sequence_dict = {}
for i, seq in enumerate(sequence_list_split):
    if seq not in sequence_dict:
        sequence_dict[seq] = [val_idx[i]]
    else:
        sequence_dict[seq].append(val_idx[i])
```

```python
val_len = len(val_idx)
new_val_len = np.floor(val_len/2)
new_test_len = val_len - new_val_len
```

```python
new_val_idx_list = []
new_test_idx_list = []
# iterate over the dictionary 
for key, values in sequence_dict.items():
    for value in values:
        if len(new_val_idx_list) < new_val_len:
            new_val_idx_list.append(value)
        else:
            new_test_idx_list.append(value)
```

```python
# write the new val and test files
with open(os.path.join(CONFIGPATH, "val_val.txt"), 'w') as f:
    for i in sorted(new_val_idx_list):
        f.write(str(i).zfill(6) + "\n")
with open(os.path.join(CONFIGPATH, "val_test.txt"), 'w') as f:
    for i in sorted(new_test_idx_list):
        f.write(str(i).zfill(6) + "\n")
```
